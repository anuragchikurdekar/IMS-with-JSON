```python
user_input_ID=str(input())
user_input_quantity=int(input())
  
if user_input_quantity<=inventory[user_input_ID]["Quantity Available"]:
  print("              BILL\n")
  print("**********")
  print("Product Purchased:",inventory[user_input_ID]["Product Name"])
  print("Price:",inventory[user_input_ID]["Price"])
  print("NO. of Product(s) Purchased:",user_input_quantity)
  print("Weight:",inventory[user_input_ID]["Weight"])
  print("Best Before:",inventory[user_input_ID]["Best Before"])
  print("Packing Date:",inventory[user_input_ID]["Packing Date"])
  print("Quantity Available:",inventory[user_input_ID]["Quantity Available"])
  print("Total amount to be paid: Rs",user_input_quantity*int((inventory[user_input_ID]["Price"])[3:]))
  print("\n**********")
  inventory[user_input_ID]["Quantity Available"]=inventory[user_input_ID]["Quantity Available"]-user_input_quantity
  print(inventory[user_input_ID]["Quantity Available"])

else:
  print("Sorry.We do not have enough supply of the product you have requested.")       
```

```python
num_of_inputs=int(input())
sales={}

for i in range(1,num_of_inputs+1):
  if user_input_quantity<=inventory[user_input_ID]["Quantity Available"]:
    user_input_ID=str(input())
    user_input_quantity=int(input())
    sales[i] = {"Product ID":user_input_ID,"Product Name":inventory[user_input_ID]["Product Name"],"Price":inventory[user_input_ID]["Price"],"Quantity Purchased":user_input_quantity,"Bill AMount":"Rs "+str(user_input_quantity*int((inventory[user_input_ID]["Price"])[3:]))}
    print(sales)
    
    js=json.dumps(sales)


    fd=open("Sales.json","a")
    fd.write(js)
    fd.close()

```
